```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.columns
```

```python
tips.head(10)
```

```python
tips.sample(10)
```

```python
sns.relplot(x="total_bill", y="tip", color = 'b', data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker" ,palette = 'viridis', data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker", data=tips,palette = 'viridis');
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips,palette = 'viridis');
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips);
```

```python
sns.jointplot("total_bill", "tip", data=tips,kind='hex');
```

```python
sns.displot(data = tips, x = 'total_bill', kde=True, bins = 30);
plt.axvline(tips['total_bill'].mean(), color="k", linestyle="--");
```

```python
b = sns.boxplot(data = tips, x="total_bill")
b.axes.set_title("Title",fontsize=50)
b.set_xlabel("X Label", fontsize=30)
b.set_ylabel("Y Label",fontsize=20)
b.tick_params(labelsize=15)
```

```python
g = sns.PairGrid(data = tips, vars=['total_bill', 'tip', 'size',  'sex'], hue='sex')
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```

```python
sns.displot(data=tips, x="total_bill", hue = "sex", kind = "kde");
```

```python
sns.violinplot(x = "sex", y = "total_bill", data=tips, palette=["lightblue", "lightpink"]);
```

```python
sns.relplot(x="tip", y="total_bill", kind="line", ci="sd", data=tips);
```

```python
sns.countplot(x='sex',data=tips);
```

```python
sns.catplot(data = tips, x = "sex", y = "tip", kind = "bar");
```

```python
sns.catplot(data = tips, x = "day",y = "tip", kind = "bar");
```

```python
sns.catplot(data = tips, x = "day" , y = "tip", kind = "bar", hue="size");
```

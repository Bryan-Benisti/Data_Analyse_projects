```python
# Lets create a new DataFrame by filtering the original one

# First I create a New column called DateYear that gives me only the Year of the events dates

My_Data['DateYear'] = pd.to_datetime(My_Data['Event Date']).dt.year

My_Data.describe(include="all")
```

```python
#Now let's filter the data frame keeping only the data between 2022 to 2023 using our new column

My_Data_2022_to_2023 = My_Data[(My_Data['DateYear'] >= 2022) & (My_Data['DateYear'] <= 2023)]

#let's see the difference of rows'number

print(My_Data.shape)

print(My_Data_2022_to_2023.shape)
```

```python
# Now I want to filter again but this time I want to keep only the main event of each dates from 2022 to now. 

Main_Fight = My_Data_2022_to_2023["Event Date"].unique()  

My_Data_Mains_Events = pd.DataFrame() 

for date in Main_Fight:

    fight = My_Data_2022_to_2023.loc[My_Data_2022_to_2023["Event Date"] == date].iloc[0]  

    My_Data_Mains_Events = My_Data_Mains_Events.append(fight, ignore_index=True)  

My_Data_Mains_Events.head(51)

```

```python
# Number of fights in each event:

Nb_of_fight_grouped = My_Data_2022_to_2023.groupby("Event Name",as_index=False).agg({"Result": 'count'})

Nb_of_fight_grouped
```

```python
# Average rounds by fighter

avg_rounds_won_by_fighter = My_Data_2022_to_2023.groupby("Result", as_index=False).agg({"Round": lambda x: sum([int(n) for n in x]) / len(x)})

avg_rounds_won_by_fighter
```

```python
# create a pivot table, that shows the average number of rounds fought in each weight class for each event, with 1 digit after the point,
# And replacing "NaN" By 0

pivot_table = My_Data_2022_to_2023.pivot_table(values='Round', index=['Event Name', 'Event Date'], columns='Weight Class', aggfunc='mean')

pivot_table = pivot_table.round(1)

pivot_table = pivot_table.fillna(0)

pivot_table
```

```python
# Let's now make a Linear Regression 

Lin_Reg = smf.ols(formula = "Round ~ Time" , data = My_Data_2022_to_2023).fit()

print(Lin_Reg.summary())
```

This is a linear regression analysis that predicts the round number (dependent variable) based on the time at which the round finished (independent variable). The R-squared value of 0.717 suggests that approximately 71.7% of the variance in round number can be explained by the time at which the round finished. The Adj. R-squared value of 0.587 suggests that the model may not fit the data as well as it could, as it is less than the R-squared value. The F-statistic value of 5.494 is a measure of the overall significance of the regression and indicates that the model is significant at the p<0.05 level. The Prob (F-statistic) value of 8.50e-50 is the probability of getting an F-statistic as large as the one calculated by chance, and since it is very small, we can reject the null hypothesis that the model is not significant.
